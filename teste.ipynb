# **Tipos de Vari√°veis em An√°lise de Dados**  

Em estat√≠stica e ci√™ncia de dados, as vari√°veis podem ser classificadas em diferentes tipos com base na natureza dos dados que representam.  

## **1. Vari√°veis Qualitativas (Categ√≥ricas)**  
S√£o aquelas que expressam categorias ou atributos sem uma ordem num√©rica significativa.  

### **a) Nominais**  
- N√£o possuem uma ordem ou hierarquia.  
- Exemplo: G√™nero (Masculino, Feminino), Cores (Azul, Verde, Vermelho).  

### **b) Ordinais**  
- Possuem uma ordem ou hierarquia, mas sem uma diferen√ßa num√©rica bem definida entre os valores.  
- Exemplo: N√≠vel de escolaridade (Fundamental, M√©dio, Superior), Classifica√ß√£o de satisfa√ß√£o (Ruim, Regular, Bom, √ìtimo).  

---

## **2. Vari√°veis Quantitativas (Num√©ricas)**  
S√£o aquelas representadas por n√∫meros e permitem opera√ß√µes matem√°ticas.  

### **a) Discretas**  
- Assumem valores inteiros e cont√°veis.  
- Exemplo: N√∫mero de filhos, Quantidade de carros em um estacionamento.  

### **b) Cont√≠nuas**  
- Podem assumir qualquer valor dentro de um intervalo, incluindo n√∫meros decimais.  
- Exemplo: Altura (1,75m), Peso (68,5 kg), Temperatura (22,3¬∞C).  

---

## **3. Como Identificar o Tipo de Vari√°vel?**  
Para determinar o tipo de uma vari√°vel, pergunte-se:  
‚úî **Os valores s√£o categorias sem ordem?** ‚Üí **Qualitativa Nominal**  
‚úî **Os valores t√™m ordem, mas n√£o s√£o num√©ricos?** ‚Üí **Qualitativa Ordinal**  
‚úî **Os valores s√£o num√©ricos inteiros e cont√°veis?** ‚Üí **Quantitativa Discreta**  
‚úî **Os valores s√£o num√©ricos e podem ter casas decimais?** ‚Üí **Quantitativa Cont√≠nua**  

Compreender os tipos de vari√°veis √© essencial para escolher as t√©cnicas estat√≠sticas corretas e interpretar os dados corretamente. üöÄ

# **M√©dia, Mediana e Moda**  

Em estat√≠stica descritiva, **m√©dia, mediana e moda** s√£o medidas de tend√™ncia central, utilizadas para resumir um conjunto de dados.  

---

## **1. M√©dia**  

A **m√©dia aritm√©tica** √© calculada somando todos os valores e dividindo pelo n√∫mero total de elementos.  

### **F√≥rmula da M√©dia Aritm√©tica**  
\[
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
\]  

Onde:  
- \(\bar{x}\) = m√©dia  
- \(x_i\) = cada valor do conjunto de dados  
- \(n\) = n√∫mero total de elementos  

**Exemplo**: Se temos os n√∫meros **3, 5, 7**, a m√©dia √©:  
\[
\bar{x} = \frac{3 + 5 + 7}{3} = \frac{15}{3} = 5
\]  

---

## **2. Mediana**  

A **mediana** √© o valor central de um conjunto de dados ordenado.  

- Se a quantidade de elementos (\(n\)) for **√≠mpar**, a mediana √© o valor central.  
- Se \(n\) for **par**, a mediana √© a m√©dia dos dois valores centrais.  

**Exemplo (n √≠mpar)**: Para os n√∫meros **2, 4, 7, 9, 11**, a mediana √© **7**.  
**Exemplo (n par)**: Para os n√∫meros **1, 3, 5, 7, 9, 11**, a mediana √©:  
\[
\frac{5 + 7}{2} = 6
\]  

---

## **3. Moda**  

A **moda** √© o valor que mais se repete no conjunto de dados.  

- Pode haver **uma √∫nica moda (unimodal)**, **mais de uma moda (multimodal)** ou **nenhuma moda** se todos os valores forem distintos.  

**Exemplo**: Para os n√∫meros **4, 5, 6, 5, 7, 5, 8**, a moda √© **5** (pois aparece mais vezes).  

---

## **Resumo das Diferen√ßas**  

| Medida   | Descri√ß√£o | Quando Usar? |
|----------|----------|--------------|
| **M√©dia**   | Soma dos valores dividida pelo total | √ötil para dados sim√©tricos, sem valores extremos |
| **Mediana** | Valor central dos dados ordenados | Melhor para dados com outliers (valores extremos) |
| **Moda**    | Valor mais frequente | √ötil para identificar padr√µes e repeti√ß√µes |

Essas medidas ajudam a entender a distribui√ß√£o dos dados e escolher a melhor forma de analis√°-los.  


````python
media_anos_estudo = dados['Anos de Estudo'].mean()
media_renda = dados['Renda'].mean()
media_altura = dados['Altura'].mean()

mediana_anos_estudo = dados['Anos de Estudo'].median()
mediana_renda = dados['Renda'].median()
mediana_altura = dados['Altura'].median()

print(f"M√©dia Anos de Estudo: {media_anos_estudo}")
print(f"M√©dia Renda: {media_renda}")
print(f"M√©dia Altura: {media_altura}")

print(f"Mediana Anos de Estudo: {mediana_anos_estudo}")
print(f"Mediana Renda: {mediana_renda}")
print(f"Mediana Altura: {mediana_altura}")
````

# **Correla√ß√£o e Coeficiente de Determina√ß√£o (\( R^2 \))**  

## **1. O que √© Correla√ß√£o?**  

A correla√ß√£o mede o grau de associa√ß√£o entre duas vari√°veis e varia entre **-1 e 1**:  

- **\( r > 0 \)** ‚Üí Correla√ß√£o positiva (quando uma vari√°vel aumenta, a outra tamb√©m aumenta).  
- **\( r < 0 \)** ‚Üí Correla√ß√£o negativa (quando uma vari√°vel aumenta, a outra diminui).  
- **\( r = 0 \)** ‚Üí Nenhuma correla√ß√£o (as vari√°veis n√£o t√™m rela√ß√£o linear).  

---

## **2. O que √© \( R^2 \)?**  

O **coeficiente de determina√ß√£o** (\( R^2 \)) indica **quanto da varia√ß√£o de uma vari√°vel pode ser explicada pela outra**. Ele √© obtido elevando ao quadrado o coeficiente de correla√ß√£o (\( r \)):  

\[
R^2 = r^2
\]

### **Interpreta√ß√£o de \( R^2 \)**  

- **\( R^2 = 1 \)** ‚Üí A vari√°vel independente explica **100%** da varia√ß√£o da vari√°vel dependente.  
- **\( R^2 = 0.8 \)** ‚Üí A vari√°vel independente explica **80%** da varia√ß√£o.  
- **\( R^2 = 0 \)** ‚Üí Nenhuma explica√ß√£o da varia√ß√£o da vari√°vel dependente.  

‚ö†Ô∏è **Aten√ß√£o**: Um \( R^2 \) alto indica que h√° rela√ß√£o entre as vari√°veis, mas **n√£o significa que uma causa a outra**!  

---

O \( R^2 \) √© amplamente usado em **regress√£o linear** para medir a qualidade do modelo e avaliar se os dados seguem um padr√£o poss√≠vel

# Boxplot: Visualiza√ß√£o Estat√≠stica de Distribui√ß√£o

## O que √© um Boxplot?
O **Boxplot** (ou diagrama de caixa) √© uma ferramenta estat√≠stica utilizada para visualizar a **distribui√ß√£o de dados** de forma resumida. Ele destaca medidas como **mediana, quartis e valores at√≠picos**, sendo √∫til para comparar diferentes grupos.

---

## Componentes de um Boxplot
Um boxplot √© composto por:

1. **Mediana (linha dentro da caixa)**  
   - Representa o valor central dos dados (50% dos valores est√£o abaixo e 50% acima).
   
2. **Quartis (Q1 e Q3 - bordas da caixa)**  
   - **Q1 (1¬∫ quartil - 25%)**: 25% dos dados est√£o abaixo desse valor.  
   - **Q3 (3¬∫ quartil - 75%)**: 75% dos dados est√£o abaixo desse valor.  
   - A **diferen√ßa entre Q3 e Q1** √© chamada de **Intervalo Interquartil (IQR)**.

3. **Bigodes (linhas fora da caixa)**  
   - Indicam a dispers√£o dos dados sem considerar outliers.  
   - Geralmente estendem-se at√© **1,5 √ó IQR** a partir dos quartis.

4. **Outliers (pontos individuais)**  
   - Valores **fora do intervalo esperado** (abaixo de Q1 - 1,5√óIQR ou acima de Q3 + 1,5√óIQR).  
   - Podem representar erros, exce√ß√µes ou padr√µes interessantes.

# Teste T de Student  

O Teste T de Student √© um teste estat√≠stico utilizado para comparar as m√©dias de dois grupos e determinar se a diferen√ßa entre elas √© estatisticamente significativa.  

## Quando usar o Teste T?  
O teste T √© aplicado quando queremos verificar se duas amostras s√£o significativamente diferentes em rela√ß√£o a uma vari√°vel num√©rica.  

### Exemplos:  
- Comparar anos de estudo entre pessoas de diferentes grupos.  
- Comparar sal√°rios entre homens e mulheres.  
- Comparar altura entre dois grupos.  

## Tipos de Teste T  

1. **Teste T para amostras independentes**  
   - Compara dois grupos diferentes (exemplo: "anos de estudo" entre dois grupos distintos).  

2. **Teste T para amostras pareadas**  
   - Compara o mesmo grupo antes e depois de um tratamento (exemplo: comparar sal√°rios de um grupo antes e depois de um curso).  

3. **Teste T para uma amostra**  
   - Compara a m√©dia de uma amostra com um valor fixo (exemplo: verificar se a m√©dia de renda de um grupo √© diferente de um determinado valor).  

## Hip√≥teses do Teste T  

- **Hip√≥tese nula (H‚ÇÄ):** N√£o h√° diferen√ßa significativa entre as m√©dias.  
- **Hip√≥tese alternativa (H‚ÇÅ):** H√° uma diferen√ßa significativa entre as m√©dias.  

Se o p-valor for menor que 0.05, rejeitamos H‚ÇÄ e consideramos que h√° diferen√ßa significativa.  

  


````python
from scipy.stats import ttest_ind

# Filtrando os dados para pessoas Negras (Preta)
dados_negras = dados[dados['Cor'] == 'Preta']
dados_nao_negras = dados[dados['Cor'] != 'Preta']

# Realizando o Teste T de Student
t_stat, p_value = ttest_ind(dados_negras['Anos de Estudo'], dados_nao_negras['Anos de Estudo'], equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Rejeitamos a hip√≥tese nula: H√° diferen√ßa significativa nos anos de estudo entre pessoas Negras e n√£o Negras.")
else:
    print("N√£o rejeitamos a hip√≥tese nula: N√£o h√° diferen√ßa significativa nos anos de estudo entre pessoas Negras e n√£o Negras.")
````

````python
import numpy as np
from scipy.stats import t

import matplotlib.pyplot as plt

# Par√¢metros para a distribui√ß√£o T de Student
df = len(dados_negras) + len(dados_nao_negras) - 2  # graus de liberdade
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

# Plotando a distribui√ß√£o T de Student
plt.figure(figsize=(12, 8))
plt.plot(x, y, label=f'T-distribution (df={df})', color='royalblue')
plt.axvline(t_stat, color='red', linestyle='--', label=f'T-statistic = {t_stat:.2f}')
plt.fill_between(x, y, where=(x >= t_stat), color='red', alpha=0.3)
plt.fill_between(x, y, where=(x <= -t_stat), color='red', alpha=0.3)
plt.title('Distribui√ß√£o T de Student')
plt.xlabel('Valores de T')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
````

# Introdu√ß√£o ao Machine Learning  

Machine Learning (ML) √© um campo da intelig√™ncia artificial que permite que sistemas aprendam padr√µes a partir de dados e tomem decis√µes sem serem explicitamente programados. Os algoritmos de ML podem ser divididos em tr√™s categorias principais:  

- **Aprendizado Supervisionado:** O modelo aprende a partir de dados rotulados, onde h√° uma rela√ß√£o entre as entradas e as sa√≠das desejadas.  
- **Aprendizado N√£o Supervisionado:** O modelo identifica padr√µes e estruturas nos dados sem r√≥tulos pr√©-definidos.  
- **Aprendizado por Refor√ßo:** O modelo aprende por meio de intera√ß√µes com o ambiente, recebendo recompensas ou penalidades com base em suas a√ß√µes.  

Dentre os diversos algoritmos utilizados no aprendizado supervisionado, um dos mais poderosos e populares √© o **Random Forest**.  

# Random Forest  

## O que √© Random Forest?  
Random Forest √© um algoritmo de aprendizado supervisionado baseado em √°rvores de decis√£o. Ele constr√≥i diversas √°rvores de decis√£o durante o treinamento e combina seus resultados para melhorar a precis√£o e reduzir o risco de overfitting.  

## Como Funciona?  

O Random Forest segue uma abordagem conhecida como **ensemble learning**, onde v√°rios modelos s√£o combinados para obter previs√µes mais robustas. Seu funcionamento ocorre em tr√™s etapas principais:  

1. **Cria√ß√£o de m√∫ltiplas √°rvores de decis√£o:** O algoritmo seleciona subconjuntos aleat√≥rios dos dados de treinamento e treina uma √°rvore de decis√£o para cada um deles. Esse processo √© chamado de **Bootstrap Sampling**.  
2. **Predi√ß√£o por cada √°rvore:** Cada √°rvore faz uma previs√£o baseada nos dados que recebeu.  
3. **Combina√ß√£o dos resultados:** Para problemas de **classifica√ß√£o**, o resultado final √© decidido por vota√ß√£o da maioria. Para **regress√£o**, a m√©dia das previs√µes das √°rvores √© utilizada.  

## Vantagens  

- Reduz o overfitting ao combinar m√∫ltiplas √°rvores.  
- Funciona bem com grandes volumes de dados e problemas complexos.  
- Pode lidar com vari√°veis categ√≥ricas e num√©ricas sem a necessidade de pr√©-processamento intenso.  
- Permite calcular a import√¢ncia das vari√°veis, ajudando na interpreta√ß√£o dos dados.  

## Desvantagens  

- Pode ser mais lento do que uma √∫nica √°rvore de decis√£o, pois cria m√∫ltiplos modelos.  
- Dif√≠cil de interpretar quando h√° um grande n√∫mero de √°rvores, tornando-o menos transparente do que m√©todos mais simples.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X = dados.drop(columns=['Anos de Estudo'])
y = dados['Anos de Estudo']

X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf = RandomForestRegressor(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
````
